# Importing modules

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

print('Modules imported')
```

## Importing covid19 dataset

```python
corona_dataset_csv = pd.read_csv('Datasets/time_series_covid19_confirmed_global.csv')
```

```python
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

## Deleting useless columns

```python
corona_dataset_csv.drop(['Lat', 'Long'], axis = 1, inplace = True)
```

```python
corona_dataset_csv.head(10)
```

## Aggregating rows by country

```python
corona_dataset_aggregated = corona_dataset_csv.groupby('Country/Region').sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape
```

## Visualizing data for country

```python
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

## Calculating a good measure

```python
corona_dataset_aggregated.loc['China'].plot()
```

## Calculating the first derivative of the curve

```python
corona_dataset_aggregated.loc['China'].diff().plot()
```

## Calculating maximum infection rate for China

```python
corona_dataset_aggregated.loc['China'].diff().max()
```

## Calculating maximum infection rate for all countries

```python
countries = list(corona_dataset_aggregated.index)
maximum_infection_rates = []
for country in countries:
    maximum_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated['Maximum Infection Rate'] = maximum_infection_rates
```

```python
corona_dataset_aggregated.head()
```

## New dataframe with only needed column

```python
corona_data = pd.DataFrame(corona_dataset_aggregated['Maximum Infection Rate'])
```

```python
corona_data.head()
```

## Importing happiness dataset

```python
world_happiness_report = pd.read_csv('Datasets/World Happiness Report.csv')
world_happiness_report.head()
```

```python
world_happiness_report.shape
```

## Removing useless columns

```python
columns_to_drop = ['Happiness Rank', 'Happiness Score', 'Generosity', 'Region']
world_happiness_report.drop(columns_to_drop, axis = 1, inplace = True)
```

```python
world_happiness_report.head()
```

## Changing the indices of the dataframe

```python
world_happiness_report.set_index(['Country'], inplace = True)
world_happiness_report.head()
```

## Joining datasets

### Corona dataset:

```python
corona_data.head()
```

### World Happiness Dataset:

```python
world_happiness_report.head()
```

```python
data = world_happiness_report.join(corona_data).copy()
data.head()
```

### Correlation Matrix

```python
data.corr()
```
